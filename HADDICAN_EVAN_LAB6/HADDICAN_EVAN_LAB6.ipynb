```python
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import csv
import time
```

```python
PATH = Service("C:/Users/evanhaddican/Documents/HARP151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

bing = driver.get("https://www.binghamton-ny.gov/home")
menu = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/a")
submenu = driver.find_element(By.XPATH, "//*[@id='dropdownrootitem3']/div/div/ul[1]/li/a")


actions = ActionChains(driver)
actions.move_to_element(menu)
actions.click(menu)
time.sleep(5)
actions.click(submenu)
actions.perform()

csv_file = open("bingdata.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["JOB","TYPE","APPLICATION DEADLINE","SALARY"])

try: 
    personnel_xpath = "//*[@id='leftNav_2_0_127']/ul/li/ul/li[14]/a[2]"
    employment_xpath = "//*[@id='leftNav_2_0_127']/ul/li/ul/li[14]/ul/li/a"
    personnel = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, personnel_xpath)))
    personnel.click()
    employment = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, employment_xpath)))
    employment.click()
    
    rows = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@id='ColumnUserControl4']/div[2]/table" )))
    whole_table = rows.find_elements(By.TAG_NAME, "tr")
    counter = 0
    for i in whole_table:
        if counter > 0:
            job = i.find_element(By.XPATH, f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{counter}]/td[1]/a').text
            type_job = i.find_element(By.XPATH, f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{counter}]/td[2]').text
            application_deadline = i.find_element(By.XPATH, f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{counter}]/td[3]').text
            salary = i.find_element(By.XPATH, f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{counter}]/td[4]').text
        else:
            job = "-"
            type_job = "-"
            application_deadline = "-"
            salary = "-"
        counter += 1
        csv_writer.writerow([job,type_job,application_deadline,salary])
finally:
        driver.quit()
        csv_file.close()
    
sheet = pd.read_csv("bingdata.csv")
sheet
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
import csv
from bs4 import BeautifulSoup
import pandas as pd
```

```python
PATH = Service("C:/Users/evanhaddican/Documents/HARP151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

#open the website
driver.get("https://techwithtim.net")

#search for Python Programming
link = driver.find_element(By.LINK_TEXT,"Python Programming")
#click the link
link.click()

try:
    element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials")))
    element.click()
    #click link
    
    element = WebDriverWait(driver, 10).until(EC.presence_of_element((By.ID, "sow-button-19310003")))
    element.click()
    #click link
    
    driver.back()
    driver.back()
    driver.back()
    driver.forward()
    driver.forward()
    #going forward and back
except:
    driver.quit()
    
    
```

```python
new_csv = open("salaries.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(new_csv)
csv_writer.writerow(["Salaries"])

PATH = Service("C:/Users/evanhaddican/Documents/HARP151/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

driver.get("https://hoopshype.com/")

link = driver.find_element(By.ID,"menu-item-458969")
link.click()

page = driver.find_element(By.XPATH,"/html/body/div[3]/div[2]/div[2]/div[2]/div/div[1]/div/div[3]/ul/li[2]/a")
page.click()

try:
    main = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID,"main")))
    rows = main.find_elements(By.TAG_NAME,"tr")
    for row in rows[1:]:
        salary = row.find_element(By.CLASS_NAME, "hh-salaries-sorted").text
        csv_writer.writerow([salary])
finally:
    driver.quit()
    new_csv.close()

pd.read_csv("salaries.csv")
```

```python
2.
new_csv = open("library.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(new_csv)
csv_writer.writerow(["Title", "Description", "Hyperlink"])

page = 1
num = 1

while page != 5:
    library = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    puos = BeautifulSoup(library, "lxml")
    
    for element in puos.find_all("div", class_="item-description"):
        title = element.find("span", class_="item-description-title").text.strip("Â")
        try:
            description = element.find("span", class_="item-description-abstract").text.strip("Â")
        except AttributeError:
            description = "None"
        hyperlink = element.find("a").get('href')
        print(f"{num}. {title} - - {description} - - {hyperlink}")
        csv_writer.writerow([title, description, hyperlink])
        
    page += 1
    num += 1
new_csv.close()
```

```python
from tkinter import *
import requests
import random
```

```python
root = Tk()
root.title("Trivia")
root.geometry("800x800")

url = "https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy"
content = requests.get(url).json()

frame = Frame(root)
frame = frame.pack()

score = 0

radio = StringVar()
radio.set("")

def newQuestion():
    frame = Frame(root)
    frame.pack()
    
    label = Label(frame)
    label.pack(anchor=S)    
    
    
    def selection():
        global score
        select = radio.get()
        if select == correct:
            score += 1
            label.config(text=f"You're right! Your score is {score}")
        else:
            score -= 1
            label.config(text=f"You're wrong! Your score is {score}")
    
    for i in range(0,1):
        x = random.randint(0,8)
        y = random.randint(0,3)
    
        q = content[x]
        incorrect = q["incorrectAnswers"]
        correct = q["correctAnswer"]
        incorrect.insert(y, correct)
        question = q["question"]
    
    Label(frame, text=question, font=('Aerial 13')).pack(anchor=N)

    r1 = Radiobutton(frame, text=incorrect[0], variable=radio, value=incorrect[0], command=selection)
    r1.pack()

    r2 = Radiobutton(frame, text=incorrect[1], variable=radio, value=incorrect[1], command=selection)
    r2.pack()

    r3 = Radiobutton(frame, text=incorrect[2], variable=radio, value=incorrect[2], command=selection)
    r3.pack()
                  
    r4 = Radiobutton(frame, text=incorrect[3], variable=radio, value=incorrect[3], command=selection)
    r4.pack()
        
    def destroy():
        frame.destroy()
        
    
    choose_button = Button(frame, text="Next question", command=lambda: [destroy(),newQuestion()])
    choose_button.pack()

    

newQuestion()

                  
root.mainloop()
```
